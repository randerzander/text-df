DataFrame based analytics are quite popular for structured/tabular data.

Pandas, Dask, Spark, etc. can easily read a variety of formats:

* Parquet
* ORC
* CSV
* JSON
* XML

It's great when applications support these common interchange formats. But, many don't.

This means using DataFrame tools to analyze or otherwise process data can become an integration challenge.
